# Imports 

```python
from math import sqrt
import networkx as nx
import matplotlib.pyplot as plt
```

# Make an easy 1D chain

```python
G = nx.Graph(directed=False)
    
size=6    
    
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(4,5)
G.add_edge(5,6)
G.add_edge(1,6)


pos={}

for i in range(1,size+1): 
    
    # first cell
    
    pos[i]=[i,0]
    
    
print(pos)
    

nx.draw(G, pos=pos)
```

## Make a wire Hexagon that we will use to build our Lattice

```python
Hex = nx.Graph(directed=False)

size=6

for i in range(size):
    Hex.add_node(i+1, size=100, weight=0.1)

Hex.add_edge(1,2)
Hex.add_edge(2,3)
Hex.add_edge(3,4)
Hex.add_edge(4,5)
Hex.add_edge(5,6)
Hex.add_edge(1,6)
    
pos={1:[1,0], 2:[0.5,0.866], 3:[-0.5,0.866], 4:[-1,0], 5:[-0.5, -0.866], 6:[0.5, -0.866]}

nx.draw(Hex, pos=pos, node_size=0.1, alpha=0.7, node_color='black')
```

# We can build our array out using lattice vectors 

```python

num_of_hexes=10

vector_1=[0,1]
for i in range(num_of_hexes):
    
    vector_combo=[1.5,sqrt(3)/2]
    
    vx=vector_combo[0]*i
    vy=vector_combo[1]*i
    
    pos1={1:[1+vx,0+vy], 2:[0.5+vx,0.866+vy], 3:[-0.5+vx,0.866+vy], 
         4:[-1+vx,0+vy], 5:[-0.5+vx, -0.866+vy], 6:[0.5+vx, -0.866+vy]}
    nx.draw(Hex, pos=pos1, node_size=0.001, alpha=0.7, node_color='black')
    
    
for i in range(num_of_hexes): 
    
    vector_combo=[0,sqrt(3)]
    
    vx=vector_combo[0]*i
    vy=vector_combo[1]*i
    
    pos2={1:[1+vx,0+vy], 2:[0.5+vx,0.866+vy], 3:[-0.5+vx,0.866+vy], 
         4:[-1+vx,0+vy], 5:[-0.5+vx, -0.866+vy], 6:[0.5+vx, -0.866+vy]}
    nx.draw(Hex, pos=pos2, node_size=0.001, alpha=0.7, node_color='black')
```

# Include all of the possible combos and clean up the output by putting the drawing onto a matplotlib figure where we can edit the figsize by adding an axis of desired size

```python
num_of_hexes=5

fig=plt.figure()
fig.add_axes([0,0,1,2])


# draw slanted parts
vec1=[0,sqrt(3)]
vec2=[1.5,sqrt(3)/2]


for i in range(num_of_hexes): 
    
    for j in range(num_of_hexes): 
        

        
        vector_combo=[vec1[0]*i + vec2[0]*j, vec1[1]*i + vec2[1]*j]
        
        vx=vector_combo[0]
        vy=vector_combo[1]
        
            
        pos={1:[1+vx,0+vy], 2:[0.5+vx,0.866+vy], 3:[-0.5+vx,0.866+vy], 
         4:[-1+vx,0+vy], 5:[-0.5+vx, -0.866+vy], 6:[0.5+vx, -0.866+vy]}
        
        
        nx.draw(Hex, pos=pos, node_size=0.001, alpha=0.8, node_color='black', ax=fig.axes[0])
        

```

# Adding a third Vector, (vertical inverse of vector 2), gives us even more of a honey comb

```python
num_of_hexes=5

fig=plt.figure()
fig.add_axes([0,0,2,3])


# draw slanted parts
vec1=[0,sqrt(3)]
vec2=[1.5,sqrt(3)/2]
vec3=[1.5, -sqrt(3)/2]

for i in range(num_of_hexes): 
    
    for j in range(num_of_hexes): 
        

        for k in range(num_of_hexes):
            
            vector_combo=[vec1[0]*i + vec2[0]*j + vec3[0]*k, vec1[1]*i + vec2[1]*j + vec3[1]*k]

            vx=vector_combo[0]
            vy=vector_combo[1]


            pos={1:[1+vx,0+vy], 2:[0.5+vx,0.866+vy], 3:[-0.5+vx,0.866+vy], 
             4:[-1+vx,0+vy], 5:[-0.5+vx, -0.866+vy], 6:[0.5+vx, -0.866+vy]}


            nx.draw(Hex, pos=pos, node_size=0.001, alpha=0.8, node_color='black', ax=fig.axes[0])
        
```
